## 5.3 패키지 버전 이해하기

**Semver 버저닝**  
version a.b.c:
- a: major 버전, 0이면 초기 개발중, 1부터 정식 버전, major 버전은 하위호환이 안될정도로 패키지 내용이 수정되었을 때 올린다.
- b: miner 버전, 하위호환이 되는 기능 업데이트를 할 때 올린다.
- c: patch 버전, 기존 기능에 문제가 있어서 수정했을 때 올린다.


**^의 의미**  
~~~
"dependencies": {
    "body-parser": "^1.19.0",
    "cookie-parser": "^1.4.5",
    "express": "^4.17.1"
  },
  "devDependencies": {
    "nodemon": "^2.0.4",
    "rimraf": "^3.0.2"
  }
  ~~~
버전 앞에 붙은 업데이트가 일어나도 ^는 major 버전을 유지해야함을 의미한다.  
- ^: major 버전까지 유지하겠다
- ~: minor 버전까지 유지하겠다
- 아무표시없음: patch 버전까지 유지하겠다


## 5.4 기타 npm 명령어

- cpm CLI docs: https://docs.npmjs.com/cli-documentation/cli  
　  
   
 - npm outdated: 업데이트할 수 있는 패키지가 있는지 확인 가능
 - npm uninstall 패키지명: 패키지 삭제
 - npm search 검색어: npm 패키지 검색
 - npm info 패키지명: 패키지의 세부 정보 파악 가능
 - npm adduser: npm에 로그인을 하기 위한 명령어
 - npm whoami: 현재 사용자가 누구인지 알려줌
 - npm logout: 로그인한 계정을 로그아웃
 - npm version 버전: package.json의 버전을 올림
 - npm deprecate [패키지명][버전][메시지]: 패키지를 설치할 때 경고 메세지를 띄움
 - npm publish: 자신이 만든 패키지를 배포
 - npm unpublish: 자신이 만든 패키지 배포 중단(배포후 24시간 이내에만 가능)
